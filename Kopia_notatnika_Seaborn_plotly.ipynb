# Plotly basics

### Imports

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)

import seaborn as sns

import plotly.graph_objects as go
import plotly.express as px

from PIL import Image
```

## Dataset

```python
df = sns.load_dataset("mpg")
```

```python
df.head()
```

# Scatter plots

```python
# scatter plot dwóch wybranych cech
fig = px.scatter(df, x="horsepower", y="mpg")
fig.show()
```

```python
# dodanie informacji o nazwie oraz pozostałych danych do tooltipu
fig = px.scatter(df, x="horsepower", y="mpg", hover_name="name", hover_data=["cylinders", "mpg", "horsepower", "weight"])
fig.show()
```

```python
# dodanie poziomej linii widocznej na legendzie
fig = px.scatter(df, x="horsepower", y="mpg")
fig.add_trace(
    go.Scatter(x=[30,250], y=[20,20], mode="lines", name="Low", line = dict(width = 1,color = 'green'))
)
fig.add_trace(
    go.Scatter(x=[30,250], y=[40,40], mode="lines", name="High", line = dict(width = 1,color = 'red'))
)
fig.show()
```

```python
# dodanie poziomej linii niewidocznej na legendzie
fig = px.scatter(df, x="horsepower", y="mpg", trendline="ols")
fig.add_hline(y=30)
fig.show()
```

# Subplots

```python
from plotly.subplots import make_subplots
```

```python
fig = make_subplots(rows=1, cols=2, column_widths=[0.5, 0.5], shared_yaxes=True, subplot_titles=("Plot 1", "Plot 2"))
fig.add_trace(go.Scatter(x=df.horsepower, y=df.mpg, name="HP vs MPG"), row=1, col=1)
fig.add_trace(go.Scatter(x=df.horsepower, y=df.acceleration, name="HP vs 0-60"), row=1, col=2)

# change to scatter
fig.update_traces(dict(mode="markers"))

# remove legend
fig.update_layout(showlegend=False)
fig.show()
```

# Distribution plots

```python
df = sns.load_dataset("mpg")
```

```python
fig = px.histogram(df, x="mpg", color="cylinders",
                   marginal="box", opacity=0.5,
                   hover_data=df.columns, width=800, height=800)
fig.show()
```

```python
fig = px.histogram(df, x="mpg", y="horsepower", color="cylinders",
                   marginal="box", opacity=0.5, histfunc="avg",
                   hover_data=df.columns, width=800, height=800)
fig.show()
```
